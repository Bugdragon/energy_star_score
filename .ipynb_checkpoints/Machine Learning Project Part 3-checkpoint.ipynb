## Imports

```python
import pandas as pd
import numpy as np

# 去除切片复制警告
pd.options.mode.chained_assignment = None
pd.set_option('display.max_columns', 60)

import matplotlib.pyplot as plt
%matplotlib inline

plt.rcParams['font.size'] = 24

from IPython.core.pylabtools import figsize

import seaborn as sns
sns.set(font_scale = 2)

# 插补缺失值
from sklearn.preprocessing import Imputer, MinMaxScaler

# 机器学习
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import GradientBoostingRegressor

from sklearn import tree

# LIME解释预测
import lime
import lime.lime_tabular
```

## Read in Data

```python
# 将数据读取到dataframes
train_features = pd.read_csv('data/training_features.csv')
test_features = pd.read_csv('data/testing_features.csv')
train_labels = pd.read_csv('data/training_labels.csv')
test_labels = pd.read_csv('data/testing_labels.csv')
```

## Recreate Final Model

```python
# 创建imputer用以中值填充
imputer = Imputer(strategy='median')
# 在训练特征集上训练
imputer.fit(train_features)
# 转换训练集和测试集
X = imputer.transform(train_features)
X_test = imputer.transform(test_features)
# 将y转换成一维数组（向量）
y = np.array(train_labels).reshape((-1, ))
y_test = np.array(test_labels).reshape((-1, ))
```

```python
# MAE
def mae(y_true, y_pred):
    return np.mean(abs(y_true - y_pred))
```

```python
model = GradientBoostingRegressor(loss = 'lad',
                                  n_estimators = 800,
                                  max_depth = 5,
                                  min_samples_leaf = 6,
                                  min_samples_split = 6,
                                  max_features = None,
                                  random_state = 42)
model.fit(X, y)
```

```python
# 模型预测
model_pred = model.predict(X_test)

print('最终的模型在测试集上的性能：MAE = %0.4f' % mae(y_test, model_pred))
```

# Interprete the Model

解释模型的方法：
* 特征重要性
* 局部可理解的与模型无关的解释(LIME,Locally Interpretable Model-agnostic Explainer)
* 检查整体中的单个决策树

## Feature Importances

```python
# 将特征重要性提取到dataframe
feature_results = pd.DataFrame({'feature' : list(train_features.columns),
                                'importance' : model.feature_importances_})
# 输出top-10重要特征
feature_results = feature_results.sort_values('importance', ascending=False).reset_index(drop=True)
feature_results.head(10)
```

**Site EUI (kBtu/ft²)**和**Weather Normalized Site Electricity Intensity (kWh/ft²)**是最重要的两个特征。

```python
figsize(12, 10)
plt.style.use('fivethirtyeight')

feature_results.loc[:9, :].plot(x = 'feature',
                                y = 'importance',
                                edgecolor = 'k',
                                kind = 'barh',
                                color = 'blue')
plt.xlabel('Relative Importance', size=20)
plt.ylabel('')
plt.title('Feature Importances from Random Forest', size=30)
```

## Use Feature Importances for Feature Selection

### 线性回归

```python
# 提取特征
most_important_features = feature_results['feature'][:10]
# 提取索引
indices = [list(train_features.columns).index(x) for x in most_important_features]
# 保留最重要的特征
X_reduced = X[:, indices]
X_test_reduced = X_test[:, indices]

print('最重要的训练集特征形状：', X_reduced.shape)
print('最重要的测试集特征形状：', X_test_reduced.shape)
```

```python
lr = LinearRegression()
# 全特征
lr.fit(X, y)
lr_full_pred = lr.predict(X_test)
# 重要特征
lr.fit(X_reduced, y)
lr_reduced_pred = lr.predict(X_test_reduced)
# 输出结果
print('线性回归全特征结果：MAE = %0.4f' % mae(y_test, lr_full_pred))
print('线性回归重要特征结果：MAE = %0.4f' % mae(y_test, lr_reduced_pred))
```

* 减少特征并没有改善线性回归结果。
* 重要性低的特征中的额外信息可以提高性能。

### 梯度提升回归

```python
model.fit(X_reduced, y)
model_reduced_pred = model.predict(X_test_reduced)

print('最终的梯度提升回归全特征结果：MAE = %0.4f' % mae(y_test, model_pred))
print('最终的梯度提升回归重要特征结果：MAE = %0.4f' % mae(y_test, model_reduced_pred))
```

* 特征数量减少，模型结果稍差。
* 具有足够性能的最简单模型训练更快，且更易解释。

## Locally Interpretable Model-agnostic Explanations

使用LIME来解释模型所做的个别预测。
LIME是一项相对较新的工作，旨在通过使用线性模型近似预测周围的区域来展示机器学习模型的思考方式。

```python
# 残差
residuals = abs(model_reduced_pred - y_test)
# 提取最差预测和最好预测
wrong = X_test_reduced[np.argmax(residuals), :]
right = X_test_reduced[np.argmin(residuals), :]
```

```python
# 创建lime解释器对象
explainer = lime.lime_tabular.LimeTabularExplainer(training_data = X_reduced,
                                                   mode = 'regression',
                                                   training_labels = y,
                                                   feature_names = list(most_important_features))
```

```python
# 显示错误实例的预测值和真值
print('预测值： %0.4f' % model.predict(wrong.reshape(1, -1)))
print('真实值： %0.4f' % y_test[np.argmax(residuals)])
# 错误预测的解释
wrong_exp = explainer.explain_instance(data_row = wrong,
                                       predict_fn = model.predict)
# 绘制预测解释
wrong_exp.as_pyplot_figure()
plt.xlabel('Effect on Prediction', size=22)
plt.ylabel('')
plt.title('Explanation of Prediction', size=28)
```

```python
wrong_exp.show_in_notebook(show_predicted_value=False)
```

* **Site EUI** (kBtu/ft²)： 144.90 **> 95.9**
* **EDA结论**：Site EUI越大，分数越低。

```python
# 显示正确实例的预测值和真值
print('预测值： %0.4f' % model.predict(right.reshape(1, -1)))
print('真值： %0.4f' % y_test[np.argmax(residuals)])
# 正确预测的解释
right_exp = explainer.explain_instance(right,
                                       model.predict,
                                       num_features=10)
# 绘制预测解释
right_exp.as_pyplot_figure()
plt.xlabel('Effect on Prediction', size=22)
plt.ylabel('')
plt.title('Explanation of Prediction', size=28)
```

```python
right_exp.show_in_notebook(show_predicted_value=False)
```

## Examining a Single Decision Tree

* 基于树的集合,查看任何单个估计器。
* 最终模型由800个决策树组成。 

```python
# 提取单个树
single_tree = model.estimators_[105][0]
# 保存为.dot,树的最大深度5
tree.export_graphviz(single_tree,
                     out_file = 'images/tree.dot',
                     rounded = True,
                     feature_names = most_important_features,
                     filled = True)
single_tree
```

**将.dot转成.png:**
* 安装graphviz(https://www.graphviz.org/)
* 命令行：dot -Tpng images/tree.dot -o images/tree.png

```python
# 限制树的最大深度为3
tree.export_graphviz(single_tree,
                     out_file = 'images/tree_small.dot',
                     rounded = True,
                     feature_names = most_important_features,
                     filled = True,
                     max_depth = 3)
```

每个节点有**4个**不同的**信息**：
* question:根据它的答案决定向左或右子树移动
* friedman_mse：给定节点中所有示例的错误度量
* samples：节点中的示例数
* value：节点中所有示例的目标预测
